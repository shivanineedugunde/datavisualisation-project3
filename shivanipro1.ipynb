```python
# Define the label_mood function
def label_mood(valence):
    if valence >= 0.6:
        return 'Happy'
    elif valence <= 0.4:
        return 'Sad'
    else:
        return 'Neutral'
```

```python
# Apply the label_mood function to create a new column "Mood"
df['Mood'] = df['valence'].apply(label_mood)

# Count the number of songs in each mood category
mood_counts = df['Mood'].value_counts()
```

```python
# Create a pie chart
plt.figure(figsize=(8, 6))
plt.pie(mood_counts, labels=mood_counts.index, autopct='%1.1f%%', startangle=140, colors=['skyblue', 'lightcoral', 'lightgreen'])
plt.title('Distribution of Songs by Mood')
plt.show()
```
